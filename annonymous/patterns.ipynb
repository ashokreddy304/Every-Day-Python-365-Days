```python
n = 4
for i in range(n):
    for j in range(n):
        # Check if we are on the boundary (first/last row or first/last column)
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    # This print() MUST be outside the inner loop to start a new row
    print()
```

```python
s = 'learning python is very easy Python'
l1 = []
for i in range(-1,-(len(s)+1),-1):
    l = s[i]
    l1.append(l)
print("".join(l1))
```

```python
s = 'learning python is very easy Python'
l = s.split()
l1 = []
i = len(l)-1

while i>=0:
    l1.append(l[i])
    i = i-1
output = " ".join(l1)
print(output)
```
