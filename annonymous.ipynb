```python
a = [10,20,30,40,50,60]
b = [5,20,45,55,65]

i = 0
j = 0
result = []

while i < len(a) and j < len(b):
    if a[i] < b[j]:
        result.append(a[i])
        i = i + 1
    else:
        result.append(b[j])
        j = j + 1

# Add remaining elements from a
while i < len(a):
    result.append(a[i])
    i = i + 1

# Add remaining elements from b
while j < len(b):
    result.append(b[j])
    j = j + 1

print(result)
```
